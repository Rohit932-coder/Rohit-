### Problem 3: While Loop

Write a Python program that asks the user to enter a number. If the number is not between 1 and 10 (inclusive), keep asking until a valid number is entered. Once a valid number is entered, print 'Valid number entered!'.

```


while True:
  number = int (input ("Enter  a number: "))
  if number >= 1 and number <= 10:
   print ("Valid number!")
   break
  else :
    print("Invalid number")

```
